### Performing model evaluation now on both baseline and slightly modified models

```python
# Doing validation for the baseline algo (non modified!)
from sklearn.metrics import f1_score, precision_score, classification_report

y_pred = np.array(course_skill_gpt_labelled['proficiencyLevel'])
y_true = np.array(course_skill_gpt_labelled['gpt_proficiencyLevel'])

# Performing encoding of labels!
label_map = {'Basic': 0, 'Intermediate': 1, 'Advanced': 2}
y_pred = np.array([label_map[l] for l in y_pred])
y_true = np.array([label_map[l] for l in y_true])

print(y_pred[:20])
print(y_true[:20])
```

```python
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)

print(f'weighted f1 score is {f1}')
print(f'weighted precision score is {precision}')
```

```python
y_pred = np.array(course_skill_gpt_labelled['proficiencyLevel'])
y_true = np.array(course_skill_gpt_labelled['gpt_proficiencyLevel'])

print(classification_report(y_true, y_pred, target_names=['Basic', 'Intermediate', 'Advanced'], labels=['Basic', 'Intermediate', 'Advanced']))
```

### Job Skill Proficiency Labelling evaluation

```python
job_skill_gpt_evaluation = pd.read_csv('./data/jobpostings_skills_with_proficiency_labels.csv')
print(job_skill_gpt_evaluation.shape)
print(job_skill_gpt_evaluation.isnull().sum())
job_skill_gpt_evaluation.head(5)
```

The thing is because i was freaking dumb enough to not do multithreading and ran into alot of time out issues...so I had to manually filter out the remaining rows that were not processed by chatgpt.

```python
remaining_rows_df = job_skill_gpt_evaluation[job_skill_gpt_evaluation['gpt_proficiencyLevel'].isna() & job_skill_gpt_evaluation['gpt_error'].isna()]
print(remaining_rows_df.shape)
remaining_rows_df.head(5)
```

```python
remaining_rows_df.isnull().sum()
```

```python
# Exporting the remaining rows
remaining_rows_df.to_csv('./data/extracted_jobs_skills_dataset_remaining.csv', index=False)
```

Now I will have to manually read in the processed 979 records that were previously not processed by chatgpt.

```python
job_skill_gpt_evaluation_remaining = pd.read_csv('./data/jobpostings_skills_with_proficiency_labels_remaining.csv')
print(job_skill_gpt_evaluation_remaining.shape)
print(job_skill_gpt_evaluation_remaining.isnull().sum())
job_skill_gpt_evaluation_remaining.head(5)
```

```python
# Filter out records that were previously processed
job_skill_gpt_evaluation = job_skill_gpt_evaluation[~(job_skill_gpt_evaluation["gpt_proficiencyLevel"].isna() & job_skill_gpt_evaluation["gpt_error"].isna())]
print(job_skill_gpt_evaluation.shape)
```

```python
job_skill_gpt_evaluation_merged = pd.concat([job_skill_gpt_evaluation, job_skill_gpt_evaluation_remaining], ignore_index=True)
print(job_skill_gpt_evaluation_merged.shape)
print(job_skill_gpt_evaluation_merged['job_id'].nunique())
```

```python
job_skill_gpt_evaluation_merged.head(5)
print(job_skill_gpt_evaluation_merged.isnull().sum())
```

```python
job_skill_gpt_evaluation_merged = job_skill_gpt_evaluation_merged.sort_values(by=['job_id', 'jobtitle'], ascending=[True, True])
job_skill_gpt_evaluation_merged.head(5)
```

```python
# Drop those rows that proficiencyLevel, gpt_proficiencyLevel is null
job_skill_gpt_evaluation_merged = job_skill_gpt_evaluation_merged.dropna(subset=['proficiencyLevel','proficiencyScore','gpt_proficiencyLevel'])
print(job_skill_gpt_evaluation_merged.shape)
print(job_skill_gpt_evaluation_merged['job_id'].nunique())
print(job_skill_gpt_evaluation_merged.isnull().sum())
```

```python
# GPT evaluated proficiency breakdown
print(job_skill_gpt_evaluation_merged['gpt_proficiencyLevel'].value_counts())
print(job_skill_gpt_evaluation_merged['gpt_proficiencyLevel'].value_counts(normalize=True))

# PL proficiency extractor breakdown
print(job_skill_gpt_evaluation_merged['proficiencyLevel'].value_counts())
print(job_skill_gpt_evaluation_merged['proficiencyLevel'].value_counts(normalize=True))
```

```python
# Performance evaluation starts here
from sklearn.metrics import f1_score, precision_score, classification_report

y_pred = np.array(job_skill_gpt_evaluation_merged['proficiencyLevel'])
y_true = np.array(job_skill_gpt_evaluation_merged['gpt_proficiencyLevel'])

# Performing encoding of labels!
label_map = {'Basic': 0, 'Intermediate': 1, 'Advanced': 2}
y_pred = np.array([label_map[l] for l in y_pred])
y_true = np.array([label_map[l] for l in y_true])

print(y_pred[:20])
print(y_true[:20])
```

```python
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)

print(f'weighted f1 score is {f1}')
print(f'weighted precision score is {precision}')
```

```python
y_pred = np.array(job_skill_gpt_evaluation_merged['proficiencyLevel'])
y_true = np.array(job_skill_gpt_evaluation_merged['gpt_proficiencyLevel'])

print(classification_report(y_true, y_pred, target_names=['Basic', 'Intermediate', 'Advanced'], labels=['Basic', 'Intermediate', 'Advanced']))
```
