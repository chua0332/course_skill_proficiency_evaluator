### Performing model evaluation now on both baseline and slightly modified models

```python
# Doing validation for the baseline algo (non modified!)
from sklearn.metrics import f1_score, precision_score, classification_report

y_pred = np.array(course_skill_gpt_labelled['proficiencyLevel'])
y_true = np.array(course_skill_gpt_labelled['gpt_proficiencyLevel'])

# Performing encoding of labels!
label_map = {'Basic': 0, 'Intermediate': 1, 'Advanced': 2}
y_pred = np.array([label_map[l] for l in y_pred])
y_true = np.array([label_map[l] for l in y_true])

print(y_pred[:20])
print(y_true[:20])
```

```python
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)

print(f'weighted f1 score is {f1}')
print(f'weighted precision score is {precision}')
```

```python
y_pred = np.array(course_skill_gpt_labelled['proficiencyLevel'])
y_true = np.array(course_skill_gpt_labelled['gpt_proficiencyLevel'])

print(classification_report(y_true, y_pred, target_names=['Basic', 'Intermediate', 'Advanced'], labels=['Basic', 'Intermediate', 'Advanced']))
```
